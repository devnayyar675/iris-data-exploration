# 1. Project Title

**Exploring and Visualizing the Iris Dataset**

# 2. Problem Statement

The objective of this project is to explore the Iris dataset to understand trends, distributions, and relationships between various features such as sepal length, sepal width, petal length, and petal width, across different species of the Iris flower.

# 3. Import Libraries
We first begin by importing the necessary python libraries for data manipulation and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style parameters
sns.set_theme(style="whitegrid")
%matplotlib inline
```

# 4. Load Dataset

We will load the dataset from the `data/iris.csv` file using pandas.

```python
df = pd.read_csv('../data/iris.csv')
df.head()
```

# 5. Data Inspection

Let's check the shape, column names, and info of the dataset.

```python
print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())

# Checking types and null values
df.info()
```

# 6. Exploratory Data Analysis

Next, we will look at the descriptive statistical analysis and formally check for missing values to ensure data integrity.

```python
display(df.describe())
print("\nMissing values per column:\n", df.isnull().sum())
```

# 7. Data Visualization

We will generate scatter plots to analyze feature relationships, histograms for distributions, and boxplots to spot outliers.

```python
# Scatter plot for Sepal Length vs Width
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species', s=100)
plt.title('Sepal Length vs Sepal Width', fontsize=14)
plt.show()
```

```python
# Scatter plot for Petal Length vs Width
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', s=100)
plt.title('Petal Length vs Petal Width', fontsize=14)
plt.show()
```

```python
# Histograms for distribution analysis
df.hist(figsize=(10,8), bins=20, edgecolor='black')
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()
```

```python
# Boxplots for outlier detection
plt.figure(figsize=(10,6))
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title('Boxplots of Iris Features', fontsize=14)
plt.show()
```

# 8. Insights and Findings

- **Feature Relationships**: Petal length and petal width showcase a strong and clear positive linear correlation. When petal length increases, petal width tends to significantly increase.
- **Species Separation**: The *Setosa* species is beautifully disjointed from *Versicolor* and *Virginica*, particularly when looking at petal dimensions—this makes it easily classifiable.
- **Distribution Patterns**: Petal lengths act bimodal. There is a cluster of small measurements (acting as Setosa) and another higher cluster (Virginica and Versicolor merging).
- **Outlier Detection**: The boxplot exposes several statistical outliers pertaining to `sepal_width`, which should inform future outlier preprocessing logic on the data.

# 9. Conclusion

Through visual and statistical exploration, we can effectively delineate the distinct species classifications embedded within the Iris dataset. We determined that petal measurements serve as highly effective predictors due to low variance within the same class and notable variance between different classes.
